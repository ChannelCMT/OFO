```python
import requests
base_url="http://www.heibanke.com/lesson/crawler_ex00/"
r=requests.get(base_url)
```

```python
import re
count=1
import time
# import sys
start=time.time()
def get_next(r):
    number=re.search("<h3>.|\n*?([0-9]+).*?</h3>",r.text).group(1)
    global count
    count+=1
#     print(count)
    print(number)
    url=base_url+number
    return url
while True:
    try:
        url=get_next(r)
        r=requests.get(url)
    except:
        end_time=time.time()
        print("运行了%ds"%(end_time-start))
        break
        
```
